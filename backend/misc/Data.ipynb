```python
invo_prod = dbsession.query(Invoices, InvoiceProduct.quantity, Products,
(Products.price*InvoiceProduct.quantity).label("total")
).join(InvoiceProduct,InvoiceProduct.invoice_id==Invoices.id
).join(Products,Products.id==InvoiceProduct.product_id)

pd.read_sql(invo_prod.statement, dbsession.connection())
```

```python
invo_prod = invo_prod.subquery()

invoices_summary = db.dbsession.query(invo_prod.c.id,func.sum(invo_prod.total).label("all_invoice_total")).group_by(invo_prod.c.id)
pd.read_sql(invoices_summary.statement, db.dbsession.connection())
```
